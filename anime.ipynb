# EDA

```python
episode_counts = df['Episodes'].value_counts().head(30)

# Plotting the bar graph
plt.figure(figsize=(20,8))
episode_counts.plot(kind='bar')
plt.title('Number of Episodes ')
plt.xlabel('Episodes ')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
type_counts = df['Type'].value_counts()

plt.figure(figsize=(20,8))
type_counts.plot(kind='bar')
plt.title('Types of Anime')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
studios_counts = df['Studios'].value_counts().head(30)

plt.figure(figsize=(20,8))
studios_counts.plot(kind='barh')
plt.title('Number of Studios')
plt.xlabel('Count')
plt.ylabel('Studios')
plt.xticks(rotation=90)
plt.show()
```

```python
# prompt: make a plot for Rating  make it horizontally and in space

rating_counts = df['Rating'].value_counts()

plt.figure(figsize=(20,8))
rating_counts.plot(kind='barh')
plt.title('Rating of Anime')
plt.xlabel('Count')
plt.ylabel('Rating')
plt.xticks(rotation=90)
plt.show()
```

```python
# prompt: Duration plot horizontally

duration_counts = df['Duration'].value_counts().head(30)

plt.figure(figsize=(20,8))
duration_counts.plot(kind='barh')
plt.title('Duration of Anime')
plt.xlabel('Count')
plt.ylabel('Duration')
plt.xticks(rotation=90)
plt.show()
```

# Top 50 anime based on Popularity

```python
sorted_df = df.sort_values(by=['Popularity'], ascending=[ True])

# Select the top 40 anime based on the sorted order
top_40_anime = sorted_df.head(50)

# Print the top 40 anime
print(top_40_anime[['Name',  'Popularity']])
```

```python

ranked_df = df.sort_values(by=['Ranked'], ascending=[ True])

# Select the top 40 anime based on the sorted order
top_40_anime = ranked_df.head(40)

# Print the top 40 anime
print(top_40_anime[['Name',  'Ranked']])
```

```python
sorted_df = df.sort_values(by='Score', ascending=False)
top_40_anime = sorted_df.head(50)

# Print the top 40 anime
print(top_40_anime[['Name', 'Score']])
```

```python
sorted_df = df.sort_values(by='Completed', ascending=False)
top_40_anime = sorted_df.head(50)

# Print the top 40 anime
print(top_40_anime[['Name', 'Completed']])
```
